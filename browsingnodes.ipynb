```python
import string 
df_work['TITLE']=df_work['TITLE'].str.lower()
df_work['BULLET_POINTS']=df_work['BULLET_POINTS'].str.lower()
df_work['DESCRIPTION']=df_work['DESCRIPTION'].str.lower()
```

```python
df_work['TITLE']=df_work['TITLE'].apply(lambda x: ''.join([c for c in x if c not in string.punctuation]))
df_work['BULLET_POINTS']=df_work['BULLET_POINTS'].apply(lambda x: ''.join([c for c in x if c not in string.punctuation]))
df_work['DESCRIPTION']=df_work['DESCRIPTION'].apply(lambda x: ''.join([c for c in x if c not in string.punctuation]))
```

```python
cols=['TITLE','BULLET_POINTS','DESCRIPTION']
for x in cols:
    df_work[x]=df_work[x].apply(lambda x: x.split())
```

```python
df_work.head()
```

```python
from nltk.corpus import stopwords
import nltk 
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
for m in cols:
    df_work[m] = df_work[m].apply(lambda x: [word for word in x if word not in stop_words])    
```

```python
df_work.head()
```

```python
from nltk.stem import WordNetLemmatizer 
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()
for m in cols:
    df_work[m] = df_work[m].apply(lambda x: [lemmatizer.lemmatize(word) for word in x]) 
```

```python
from nltk import SnowballStemmer
lang="english"
steamer=SnowballStemmer(lang)
def adding_stemming(tokens):
    return [steamer.stem(token) for token in tokens]
for m in  cols:
    df_work[m]=df_work[m].apply(adding_stemming)
```

```python
import gensim.downloader as api

# Load GloVe model (choose a model that fits your needs)
glove_model = api.load("glove-wiki-gigaword-300")  # 100-dimensional GloVe model
```

```python
import numpy as np

def document_vector(tokens):
    # Filter tokens to ensure they are in the GloVe vocabulary
    valid_tokens = [token for token in tokens if token in glove_model]
    if not valid_tokens:
        # Return a zero vector if no valid tokens are found
        return np.zeros(glove_model.vector_size)
    # Average the embeddings of valid tokens
    return np.mean(glove_model[valid_tokens], axis=0)

# Assuming 'COMBINED' is a list of tokens
df_preprocessed['doc_vector'] = df_preprocessed['COMBINED'].apply(document_vector)
```

```python
# Convert list of vectors into a matrix
feature_matrix = np.vstack(df_preprocessed['doc_vector'])
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=300)
X_reduced = svd.fit_transform(feature_matrix)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(feature_matrix)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=0)
clusters = kmeans.fit_predict(X_scaled)

# Visualizing clusters can be done using PCA to reduce dimensions to 2D for plotting
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', marker='o', edgecolor='black', s=50, alpha=0.6)
plt.title('Visualization of Clustered Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization

# Assuming 'num_classes' is the number of classes in your classification task
num_classes = len(set(y_train))  # Replace 10 with the actual number of classes

model = Sequential([
    Dense(512, activation='relu', input_dim=X_scaled.shape[1]),
    BatchNormalization(),
    Dropout(0.2),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.1),
    Dense(num_classes, activation='softmax')  # Output layer for multi-class classification
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # Loss function for multi-class classification
              metrics=['accuracy'])  # Metric for evaluation

# If your target labels are not one-hot encoded, use 'sparse_categorical_crossentropy' as the loss function.
# For example: loss='sparse_categorical_crossentropy'

```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define EarlyStopping callback
callbacks = [EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)]

# Assuming you have already split your data into X_train, X_val, y_train, y_val
history = model.fit(X_train, y_train, epochs=200, batch_size=32,
                    validation_data=(X_test, y_test), callbacks=callbacks)
```
